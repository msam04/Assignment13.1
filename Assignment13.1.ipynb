```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']})


print(df)
print(df.shape)

#1) Replace nan values with proper flight number. Convert float to integer column.
df['FlightNumber'][1] = 10055
df['FlightNumber'][3] = 10075
df['FlightNumber'] = pd.to_numeric(df['FlightNumber']).round(0).astype(int)
print(type(df['FlightNumber'][0]))

#2) Split the From_To column and add to temporary dataframe
f_list = []
t_list = []
for r in range(len(df)):
    x,y = df['From_To'][r].split("_")
    f_list.append(x)
    t_list.append(y)
temp_df = pd.DataFrame({'From': f_list, 'To': t_list})
print(temp_df)

#3) Standardize capitalization of From, To columns
temp_df['From'] = temp_df['From'].str.lower()
temp_df['From'] = temp_df['From'].str.capitalize()
temp_df['To'] = temp_df['To'].str.lower()
temp_df['To'] = temp_df['To'].str.capitalize()
print(temp_df)

#4) Delete From_To and add the From and To columns
df['From'] = temp_df['From']
df['To'] = temp_df['To']
df.drop('From_To', axis=1, inplace = True)
print(df)

#5) Split and replace the recent delays column
d_values = df['RecentDelays'].tolist()
for i in range(len(d_values)):
    for j in range(3 - len(d_values[i])):
        d_values[i].append(np.nan)
        
d_values = np.asarray(d_values).reshape(5,3)
print(d_values)
#To get columns, transpose and get as rows
d_values = d_values.T
delay_1 = d_values[0]
delay_2 = d_values[1]
delay_3 = d_values[2]
print(delay_1)
delays = pd.DataFrame({'delay_1':delay_1, 'delay_2':delay_2, 'delay_3':delay_3})
print(delays)
df = pd.merge(df, delays, how = 'left', right_index=False, left_index=True)

df.drop('RecentDelays', axis = 1, inplace = True)


print(df)
        

```
